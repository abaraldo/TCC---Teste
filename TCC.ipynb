```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
pnad_dict = pd.read_excel('dict_PNAD_2016.xls', sheetname='vars')
pnad_dict['vars'].fillna(method = 'ffill', inplace = True)
pnad_dict['descr'].fillna(method = 'ffill', inplace = True)
pnad_dict.loc[(pnad_dict['vars'] == 'S01001') & (pnad_dict['Tipo'] == '3')].Tipo = int(3)
```

```python
pnad_cod = pd.read_excel('dict_PNAD_2016.xls', sheetname='cod')
col = list(pnad_cod.vars)
```

```python
pnad2016 = pd.read_csv('PNADc2016.csv')
pnad2016 = pnad2016[col]
```

```python
vars_list = list(pnad2016.columns)
vars_list.remove('VD4016')
vars_list.remove('VD4017')
vars_list.remove('VD4020')
vars_list.remove('VD4022')
vars_list.remove('V3003A')
vars_list.remove('Ano')
vars_list.remove('Trimestre')
vars_list.remove('V1030')
vars_list.remove('V2001')
vars_list.remove('V2009')
vars_list.remove('S01005')
vars_list.remove('S01021')
vars_list.remove('VD4047')
```

```python
count = 0
for i in range(len(vars_list)):
    item = vars_list[i]
    values = pnad2016[item].dropna().unique()
    dict_vars = pnad_dict.loc[pnad_dict['vars'] == item][['Tipo','dict']].set_index('Tipo').to_dict()
    for value in values:
        #value = int(value)
        pnad2016[item].loc[pnad2016[item] == value] = dict_vars['dict'][value]
    count += 1
    print('Variável {} de {}'.format(count,len(vars_list)))
```
